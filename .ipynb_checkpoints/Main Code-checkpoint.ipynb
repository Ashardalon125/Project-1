In chemistry, there exist "resonant structures" that describe various orientations of a particular chemical combination. Practically, none of these forms are ever observed to be the actual shape of the molecule, but rather, it exists as a "resonance hybrid" that is a combination between all the resonant structures.

However, in the Molecular Orbital Theory (MO), some molecules, like Benzene, can be modelled using "Electron-Withdrawing Groups and "Electron-Releasing Groups." In this model, the electrons exist as a sort of cloud over the molecule in the resonant sites. This leads to another theory; The Theory of Resonance. Based on fundamental quantum mechanics, it states that the stationary model of a molecule can be represented as the weighted sum of its component states. Because each of the resonance structures is a much simpler and well understood model, using them as the "modes" of their hybrid is a useful step. In a system where the "resonance modes" are equivalent and changing, this can lead to the hybrid having its own behaviour.

By assuming that a resonance structure can "decay" into an equivalent structure, and vice versa, we can model a system of resonance structures in order to determine what the resonance hybrid, and the true molecule, are doing.

```python
def popDecay (PopA, decayA, PopB, decayB, deltaT):
    """A general population equation that takes models simultaneous decay
    
    PopA: The population that is being focused on. This population decays into B, reducing the amount of A
    decayA:   The rate at which A converts to B
    PopB: The population that is converting to A. This population increases the amount of A as it decays.
    decayB:   The rate at which B converts to A
    deltaT:   The amount of time the conversion takes place over. Must be small for accuracy
    
    CurrentPop: Returns the current population of A"""
    
    CurrentPop = PopA - (deltaT*decayA*PopA) + (deltaT*decayB*PopB)
    return CurrentPop

#Import block
%matplotlib inline
import pylab as plt

#Variable and constant block
initPopA = 100 #Determines the amount of A to start with
initPopB = 50 #Determines the amount of B to start with
initTime = 0 #Determines at what time the model starts. Initialize to zero to match initial conditions
finalTime = 5 #Determines how long the model calculates populations for. The larger it is, the more points are needed
decayA = 1 #Decay rate of A
decayB = 1 #Decay rate of B
numPoints = 100 #Number of points of data to be generated
deltaT = (finalTime-initTime)/numPoints

#Array block
APop = [] #Holds the values of A over the t-array
BPop = [] #Holds the values of B over the t-array
time = [] #Holds the corresponding t-values

#Body Calculation
for i in range(numPoints):
    time.append(initTime+(deltaT*i))
    if (i==0):
        APop.append(popDecay(initPopA, decayA, initPopB, decayB, deltaT))
        BPop.append(popDecay(initPopB, decayB, initPopA, decayA, deltaT))
    else:
        APop.append(popDecay(APop[i-1], decayA, BPop[i-1], decayB, deltaT))
        BPop.append(popDecay(BPop[i-1], decayB, APop[i-1], decayA, deltaT))
    
plt.plot(time,APop,"k-",label="Nuclei A Approximation")
plt.plot(time,BPop,"ro",label="Nuclei B Approximation")
plt.title("Same Decay, PopA Greater Than PopB")
plt.ylabel("Nuclei")
plt.xlabel("Seconds")
plt.legend(loc="best")
plt.show()
```
